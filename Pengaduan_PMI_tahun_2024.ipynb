```python
# Analisis 1: 5 Kategori Kasus Terbanyak
# Mengambil 5 kategori teratas berdasarkan Grand Total dari data asli
top_5_categories = df_cleaned.sort_values(by='Grand Total', ascending=False).head(5)['KATEGORI KASUS'].tolist()

# Filter data yang sudah ditransformasi hanya untuk 5 kategori teratas
df_top_5 = df_long[df_long['KATEGORI KASUS'].isin(top_5_categories)]

# Visualisasi 1: Tren Bulanan untuk 5 Kasus Terbanyak
plt.figure(figsize=(15, 8))
sns.lineplot(data=df_top_5, x='BULAN', y='JUMLAH KASUS', hue='KATEGORI KASUS', marker='o')
plt.title('Tren Bulanan 5 Kategori Pengaduan PMI Terbanyak', fontsize=18)
plt.xlabel('Bulan', fontsize=12)
plt.ylabel('Jumlah Kasus', fontsize=12)
plt.xticks(rotation=45)
plt.grid(True)
plt.legend(title='Kategori Kasus')
plt.tight_layout()
plt.show()

# Analisis 2: Distribusi Kasus untuk Kategori Khusus
# Contoh: Perdagangan Orang
trade_cases = df_long[df_long['KATEGORI KASUS'] == 'Perdagangan orang']

# Visualisasi 2: Grafik Batang untuk Kasus Perdagangan Orang
plt.figure(figsize=(12, 6))
sns.barplot(data=trade_cases, x='BULAN', y='JUMLAH KASUS', palette='magma')
plt.title('Distribusi Kasus Perdagangan Orang per Bulan', fontsize=18)
plt.xlabel('Bulan', fontsize=12)
plt.ylabel('Jumlah Kasus', fontsize=12)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
